# 1. 데이터 파싱(사전구축)
## 1 - 1. 데이터 불러오기 및 전처리

```python
import pandas as pd
```

```python
data = pd.read_excel(pd.ExcelFile('과제정보(2013-2017)_미래부.과기부.xlsx'))
data.head()
```

```python
for i in range(len(data.columns)):
    print('{} : {}\n'.format(i, data.columns[i]))
```

```python
df = data[['과제명', '요약문_연구목표', '요약문_연구내용', '요약문_기대효과',
           '요약문_한글키워드', '요약문_영문키워드']]
df.tail(5)
```

```python
df.describe()
```

```python
df = df[df.과제명 != '보안과제정보']
df.describe()
```

```python
df = df[df.과제명 != '시설비']
df.describe()
```

```python
df = df[df.요약문_연구목표 != '평가관리비']
df.describe()
```

```python
df = df[df.요약문_기대효과 != '보안상 생략']
df.describe()
```

## 1-2. MeCab을 활용한 사전구축

```python
from konlpy.tag import Mecab
import collections
mecab = Mecab('c:\mecab\mecab-ko-dic')
```

```python
text = ''
for i in range(df.shape[0]):
    for j in range(df.shape[1]):
        text = text + ' ' + str(df.iloc[i][j])
```

```python
nouns = mecab.nouns(text)
print(collections.Counter(nouns))
```

```python
mecab_dic = sorted(collections.Counter(nouns).items(), key = lambda t:t[1], reverse = True)
mecab_dic = pd.DataFrame(mecab_dic)
mecab_dic.columns = ['noun', 'count']
mecab_dic.head()
```

```python
mecab_dic.to_csv('C:/Users/mydus/Desktop/KISTI_ntis/mecab_dic.csv', header=False, index=False, encoding = 'CP949')
```

## 1-2. Komoran를 활용한 사전구축

```python
from konlpy.tag import Komoran
import collections
komoran = Komoran()
```

```python
komoran_dic = collections.Counter()
for text in df.과제명:
    nouns2 = komoran.nouns("\n".join([s for s in str(text).split("\n") if s]))
    komoran_dic.update(nouns2)
```

```python
for text in df.요약문_연구목표:
    nouns2 = komoran.nouns("\n".join([s for s in str(text).split("\n") if s]))
    komoran_dic.update(nouns2)
```

```python
for text in df.요약문_연구내용:
    nouns2 = komoran.nouns("\n".join([s for s in str(text).split("\n") if s]))
    komoran_dic.update(nouns2)
```

```python
for text in df.요약문_기대효과:
    nouns2 = komoran.nouns("\n".join([s for s in str(text).split("\n") if s]))
    komoran_dic.update(nouns2)
```

```python
for text in df.요약문_한글키워드:
    nouns2 = komoran.nouns("\n".join([s for s in str(text).split("\n") if s]))
    komoran_dic.update(nouns2)
```

```python
for text in df.요약문_영문키워드:
    nouns2 = komoran.nouns("\n".join([s for s in str(text).split("\n") if s]))
    komoran_dic.update(nouns2)
```

```python
komoran_dic = sorted(komoran_dic.items(), key = lambda t:t[1], reverse = True)
komoran_dic = pd.DataFrame(komoran_dic)
komoran_dic.columns = ['noun', 'count']
komoran_dic.head()
```

```python
komoran_dic.to_csv('C:/Users/mydus/Desktop/KISTI_ntis/komoran_dic.csv', header=False, index=False, encoding = 'CP949')
```
